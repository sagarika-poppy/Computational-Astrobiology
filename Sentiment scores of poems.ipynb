```python
from pathos.multiprocessing import ProcessPool
from textblob import TextBlob
from tqdm import tqdm 
import os
```

```python
# Reading multiple poem files from my system
```

```python
directory = "/home/sagarika/belgrade/astrobio"  # replace with your directory path
files = [f for f in os.listdir(directory) if f.endswith('.txt')]

texts = []
for filename in files:
    with open(os.path.join(directory, filename), 'r') as file:
        text = file.read()
        texts.append(text)
```

```python
texts    #This is the porm i have in a text file
```

```python
def get_sentiment(text):
    from textblob import TextBlob
    blob = TextBlob(text)
    score = blob.sentiment.polarity
    return score
```

```python
lines_0=texts[0].split('\n')
lines_1=texts[1].split('\n')
lines_2=texts[2].split('\n')
```

```python
lines = lines_0 + lines_1 + lines_2
print(lines)
```

```python
#Process the process in parallel

pool = ProcessPool(nodes=4)
sentiment_scores = list(tqdm(pool.imap(get_sentiment, lines), total=len(lines)))
```

Credits: https://www.tibco.com/reference-center/what-is-massively-parallel-processing

```python
#Print the sentiment scores

print(sentiment_scores)

#Plot the sentiment scores

import matplotlib.pyplot as plt
plt.hist(sentiment_scores)
plt.show()
```

# Serial Processing

```python
#Process the poems in series

sentiment_scores = []
for lines in tqdm(lines):
    sentiment_scores.append(get_sentiment(lines))
```
